```python
Italy = pd.read_json(r"C:\Users\Nino\Desktop\events_Italy.json")
Germany = pd.read_json(r"C:\Users\Nino\Desktop\events_Germany.json")
France = pd.read_json(r"C:\Users\Nino\Desktop\events_France.json")
England = pd.read_json(r"C:\Users\Nino\Desktop\events_England.json")
Spain = pd.read_json(r"C:\Users\Nino\Desktop\events_Spain.json")
```

```python
events = pd.concat([Italy,Germany,France,England,Spain], axis=0)
```

```python
events
```

since in this RQ we are just interested in "pass" events, we filter the rows that has eventId=8

```python
pass_events = events[events.eventId == 8]
pass_events
```

Now we want to know about accurate passes so we filter rows that contain tag id 1801

```python
acc_passes = pass_events[pass_events['tags'].apply(lambda x: pd.Series([dict['id'] for dict in x]).eq(1801).any())]
```

```python
acc_passes
```

```python
ratio = ( acc_passes["playerId"].value_counts() / pass_events["playerId"].value_counts() ).sort_values(ascending=[False])
ratio
```
